```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data for preprocessing.xlsx')
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.replace('-', np.nan, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Sent dayofweek'] = df['Sent Date'].dt.day_name()
```

```python
df
```

```python
df.drop(['End date', 'Registrations','Day1 Conv.', 'Clicks', 'Month'], axis=1, inplace=True)
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df['Control Group Rev'].fillna(df['Control Group Rev'].mode(), inplace=True)
```

```python
df['Conversions'].fillna(df['Conversions'].median(), inplace=True)
```

```python
df['open_rate'] = df['Notification opens'] / df['Recipients']
average_open_rate = df['open_rate'].mean(skipna=True)
df['Notification opens'] = df['Notification opens'].fillna(df['Recipients'] * average_open_rate)
df.drop('open_rate', axis=1, inplace=True)
```

```python
df['Channel'].fillna(df['Channel'].mode()[0], inplace=True)
```

```python
df['Sent time'] = df['Sent time'].fillna(method='ffill')
```

```python
df['Voucher Name'] = df['Voucher Name'].fillna(method='ffill')
```

```python
df['Campaing Revenue'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Country'] = df['Country'].astype('category').cat.codes
df['Channel'] = df['Channel'].astype('category').cat.codes
df['Service Type'] = df['Service Type'].astype('category').cat.codes
df['Message type'] = df['Message type'].astype('category').cat.codes
df['Voucher Name'] = df['Voucher Name'].astype('category').cat.codes
df['Campaign Name'] = df['Campaign Name'].astype('category').cat.codes
df
```

```python
df['CRM Name'].describe()
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

```python
df['Control Group Rev'].replace('SAR', '', inplace=True)
df
```

```python
df['Control Group Rev'].dtype
```

```python
df['Control Group Rev'] = pd.to_numeric(df['Control Group Rev'], errors='coerce')
df['Control Group Rev'] = df['Control Group Rev'].fillna(df['Control Group Rev'].median())
```
